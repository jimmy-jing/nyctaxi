csv files are organized by month of each year and Feb 2009 is 2.2GB and around 500K rides. It seems like monthly files vary between 0.75=2.5 GB. I have decided to take 1% of each file to maintain ratio integrity.

I have downloaded the files into a folder called 'NYC_TAXI_DATA' on my S drive (larger storage drive 2.5TB). The function below will compile a sample of 1% of each file and rewrite to csv. The final csv(s) will be different based on taxi type since each taxi type has a different format and different number of columns.

```python
def taxi_sample_csv(taxi_type,year):
    #taxi types are ['yellow','green','fhv']
    
    filepath = 'S:/NYC_TAXI_DATA/'
    month = ['01','02','03','04','05','06','07','08','09','10','11','12']
    
    csv_name = taxi_type+'_'+str(year)+'.csv'
    
    for i in month:
        final_path = filepath+taxi_type+'_tripdata_'+str(year)+'-'+i+'.csv'
        df = pd.read_csv(final_path)
        print(f'{taxi_type} {year} month {i}')
#         print('total rows:',df.shape[0])
        df = df.sample(frac=0.01, random_state=50)
        print('rows extracted:',df.shape[0])
        print('='*40)
        
        #while data files are in separate folder, all compiled csvs are being written into current directory
        if i == '01':
            df.to_csv(csv_name, index=False)
            print('saved to:', csv_name,'\n')
        else:
            df.to_csv(csv_name, mode='a', header=False, index=False)
            print('added on month:',i,'\n')
        
        
```

```python
def debug_df(taxi_type, year):
    filepath = 'S:/NYC_TAXI_DATA/'
    month = ['01','02','03','04','05','06','07','08','09','10','11','12']

    list_of_df = []
    for i in month:
        final_path = filepath+taxi_type+'_tripdata_'+str(year)+'-'+i+'.csv'
        df = pd.read_csv(final_path)
        print('total rows:',df.shape[0])
        df = df.sample(frac=0.01, random_state=50)
        print('rows extracted:',df.shape[0])
        print('='*40)
        list_of_df.append(df)
        
    scan_df(list_of_df)
    return list_of_df

def scan_df(list_of_df):
    for i in range(len(list_of_df)):
        print(i,'\n')
        print(list_of_df[i].shape)
        print(list_of_df[i].dtypes)



def split_write(list_of_df, last_month,file_name):
    list_of_df[0].to_csv(file_name+'_'+str(last_month)+'.csv',index=False)
    for i in range(1,last_month):
        list_of_df[i].to_csv(file_name+'_'+str(last_month)+'.csv',mode='a', header=False, index=False)

    list_of_df[last_month].to_csv(file_name+'_'+str(last_month+1)+'.csv',index=False)
    for i in range(last_month+1,12):
        list_of_df[i].to_csv(file_name+'_'+str(last_month+1)+'.csv',mode='a', header=False, index=False)
```
